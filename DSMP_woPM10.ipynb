```python
df = df.sort_values(by='Date Time')
```

```python
df['Month'] = pd.DatetimeIndex(df['Date Time']).month
```

```python
dfs = []
for i in range(1,13):
    m = df.loc[df.Month == i]
    m['NOx'] = m['NOx'].fillna((m['NOx'].mean()))
    m['NO2'] = m['NO2'].fillna((m['NO2'].mean()))
    m['NO'] = m['NO'].fillna((m['NO'].mean()))
    m['Temperature'] = m['Temperature'].fillna((m['Temperature'].mean()))
    dfs.append(m)
    
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
#Dropping unwanted columns
cols = [5,6,7,8,9,10,11,12,13,15,16,19,20]
df.drop(df.columns[cols],axis=1,inplace=True)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("All cars vs ULEV.csv")
```

```python
df
```

```python
df['Type']=pd.to_datetime(df['Type'])
df['All Cars']=df['All Cars'].astype(int)
df['ULEV']=df['All Cars'].astype(int)
```

```python
df.set_index('Type', inplace=True)
```

```python
df
```

```python
df=df.resample('1M').agg('max')
```

```python
df=df.interpolate(method='time').reset_index()
```

```python
df.rename(columns={'Type':'Date'})
```

```python
df.to_csv("Interpolated_IncomeandPopulation.csv")
```

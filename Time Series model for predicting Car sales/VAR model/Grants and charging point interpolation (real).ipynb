```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("100 CPs and no grants.csv")
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df=df.resample('1M').agg('max')
```

```python
df=df.interpolate(method='time').reset_index()
```

```python
df['Charging Points']=df['Charging Points'].astype(int)
```

```python
df
```

```python
df_2=pd.read_csv("original Data.csv")
```

```python
df_2['Date']=pd.to_datetime(df_2['Date'])
```

```python
df_2
```

```python
final_data=pd.merge(df_2, df, how="left", on=['Date'])
```

```python
final_data.to_csv("Final data.csv",index=False)
```

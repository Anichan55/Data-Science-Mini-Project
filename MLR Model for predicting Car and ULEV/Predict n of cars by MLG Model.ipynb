```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Final data.csv")
df
```

```python
df=df.dropna()
```

```python
#separate the other attributes from the predicting attribute
x = df.drop(columns=['Date','All Cars', 'ULEV'])
#separte the predicting attribute into Y for model training 
y = df['All Cars']
Y = df['ULEV']
```

```python
x
```

```python
y
```

```python
Y
```

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle=False)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# predicting the accuracy score
score=r2_score(y_test,y_prediction)
print('r2 socre is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y_test,y_prediction)))
```

```python
# create data frame
compare_y = df[int(110*0.8):]
compare_y = compare_y['Date']
compare_y = pd.DataFrame(compare_y)
compare_y.reset_index(drop=True, inplace=True)
y_test= pd.DataFrame(y_test)
y_test.reset_index(drop=True, inplace=True)
y_prediction = pd.DataFrame(y_prediction)
y_prediction.reset_index(drop=True, inplace=True)
compare_y = pd.merge(compare_y, y_test, left_index=True, right_index=True, how="inner")
compare_y = pd.merge(compare_y, y_prediction, left_index=True, right_index=True, how="inner")
compare_y.index = compare_y['Date']
del compare_y['Date']
compare_y.columns = ['Test_All Cars', 'Predict_All Cars']
compare_y
```

```python
#plot
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize = (15,8))
plt.plot(compare_y['Test_All Cars'], label='Test_All Cars')
plt.plot(compare_y['Predict_All Cars'], label='Predict_All Cars')
plt.legend()
```
